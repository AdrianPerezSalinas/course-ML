```python
# load the required libraries

from aux_solutions import SupportVectorMachine
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = np.vstack((
    np.random.randn(50, 2) - 1,
    np.random.randn(50, 2) + 3
))
y = np.hstack((-1 * np.ones(50), np.ones(50)))



# Train SVM
svm = SupportVectorMachine()
svm.fit(X, y)

# Predict
predictions = svm.predict(X)

```

```python
# Plot decision boundary
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = model.predict(grid)
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.bwr)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr, edgecolor="k")
    plt.title("SVM Decision Boundary")
    plt.show()
    
def plot_data_points(X, y):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr, edgecolor="k")
    plt.title("SVM Data")
    plt.show()

plot_decision_boundary(X, y, svm)
```

```python
np.random.seed(0)
X = np.random.randn(100, 2)
y = 2 * np.array(np.linalg.norm(X, axis=1) > 1, dtype=int) - 1
plot_data_points(X, y)

# Train SVM
svm = SupportVectorMachine()
svm.fit(X, y)

# Predict
predictions = svm.predict(X)

plot_decision_boundary(X, y, svm)
```

```python
def feature_map(X): 
    return np.linalg.norm(X, axis = 1).reshape(len(X), 1)
```

```python
# Train SVM
svm = SupportVectorMachine(feature_map = feature_map)

costs = svm.fit(X, y, lambda_param = 0)

# Predict
predictions = svm.predict(X)



plot_decision_boundary(X, y, svm)
```
